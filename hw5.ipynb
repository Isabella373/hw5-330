````python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# pipeline: preprocessing + linear model
ridge_pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", Ridge())
])

# hyperparameter tuning over alpha (regularization strength)
param_grid = {
    "model__alpha": np.logspace(-3, 3, 13)  # 1e-3 ... 1e3
}

grid = GridSearchCV(
    ridge_pipe,
    param_grid=param_grid,
    cv=5,
    scoring="neg_mean_absolute_error",  # use MAE for skewed target
    n_jobs=-1
)

grid.fit(X_train, y_train)

best_alpha = grid.best_params_["model__alpha"]
best_cv_mae = -grid.best_score_
cv_std = grid.cv_results_["std_test_score"][grid.best_index_]

print("Best alpha:", best_alpha)
print(f"Best CV MAE: {best_cv_mae:.4f} (std: {cv_std:.4f})")
````

````python
best_model = grid.best_estimator_

y_pred_test = best_model.predict(X_test)

test_mae = mean_absolute_error(y_test, y_pred_test)
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

print("Ridge (tuned) performance on test set")
print(f"Test MAE:  {test_mae:.4f}")
print(f"Test RMSE: {test_rmse:.4f}")
````

````python
# Hyperparameter optimization (tune multiple hyperparameters)
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error

from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor

# Helper to run randomized search and report results
def run_search(name, pipe, param_distributions, n_iter=20, cv=5):
    search = RandomizedSearchCV(
        estimator=pipe,
        param_distributions=param_distributions,
        n_iter=n_iter,
        scoring="neg_mean_absolute_error",
        cv=cv,
        random_state=123,
        n_jobs=-1,
        verbose=0,
    )
    search.fit(X_train, y_train)

    best_cv_mae = -search.best_score_
    best_params = search.best_params_

    best_model = search.best_estimator_
    y_pred_test = best_model.predict(X_test)

    test_mae = mean_absolute_error(y_test, y_pred_test)
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

    print(f"\n===== {name} (RandomizedSearchCV) =====")
    print("Best params:", best_params)
    print(f"Best CV MAE: {best_cv_mae:.4f}")
    print(f"Test MAE:    {test_mae:.4f}")
    print(f"Test RMSE:   {test_rmse:.4f}")

    return {
        "model": name,
        "best_cv_mae": best_cv_mae,
        "test_mae": test_mae,
        "test_rmse": test_rmse,
        "best_params": best_params,
    }

results = []

# 1) RandomForest (tree-based ensemble) — tunes MULTIPLE hyperparameters
rf_pipe = Pipeline([
    ("preprocess", preprocessor),  # RF can work with sparse OHE
    ("model", RandomForestRegressor(random_state=123, n_jobs=-1)),
])

rf_params = {
    "model__n_estimators": [100, 200, 400],
    "model__max_depth": [None, 10, 18, 25, 35],
    "model__min_samples_split": [2, 5, 10, 20],
    "model__min_samples_leaf": [1, 2, 4, 8],
    "model__max_features": ["sqrt", "log2", 0.5, 0.8],
}

results.append(run_search("RandomForestRegressor", rf_pipe, rf_params, n_iter=25, cv=5))

# 2) HistGradientBoosting (another strong model) — also tune multiple params
hgb_pipe = Pipeline([
    ("preprocess", preprocessor_dense),  # ensures dense input
    ("model", HistGradientBoostingRegressor(random_state=123)),
])

hgb_params = {
    "model__learning_rate": [0.02, 0.05, 0.08, 0.1],
    "model__max_depth": [3, 5, 8, None],
    "model__max_leaf_nodes": [15, 31, 63, 127],
    "model__min_samples_leaf": [10, 20, 40, 80],
    "model__l2_regularization": [0.0, 0.1, 0.5, 1.0],
}

results.append(run_search("HistGradientBoostingRegressor", hgb_pipe, hgb_params, n_iter=25, cv=5))

# Summary (compare optimized results across models) ---
# sort by test MAE (lower is better)
results_sorted = sorted(results, key=lambda d: d["test_mae"])

print("\n===== Summary (sorted by Test MAE) =====")
for r in results_sorted:
    print(f"{r['model']}: CV MAE={r['best_cv_mae']:.4f} | Test MAE={r['test_mae']:.4f} | Test RMSE={r['test_rmse']:.4f}")
````
